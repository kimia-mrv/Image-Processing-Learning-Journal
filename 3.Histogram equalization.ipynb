```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import cv2
```

```python
# Define the range of x-values
x = np.linspace(-5, 5, 1000)

# Calculate the PMF
pmf = norm.pdf(x, 0, 1)

# Calculate the CMF
cmf = norm.cdf(x, 0, 1)

# Plot the PMF
plt.plot(x, pmf, label='PMF')

# Plot the CMF
plt.plot(x, cmf, label='CMF')

# Add a title and axis labels
plt.title('Gaussian PMF and CMF')
plt.xlabel('x')
plt.ylabel('Probability')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Load the image
img = cv2.imread('/Images/image3.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram of the original image
hist_orig, bins = np.histogram(img.flatten(), 256, [0, 256])

# Apply histogram equalization
img_eq = cv2.equalizeHist(img)

# Calculate the histogram of the equalized image
hist_eq, bins = np.histogram(img_eq.flatten(), 256, [0, 256])

# Apply histogram matching transform
ref_img = cv2.imread('reference_image.jpg', cv2.IMREAD_GRAYSCALE)
match = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_match = match.apply(img)

# Calculate the histogram of the matched image
hist_match, bins = np.histogram(img_match.flatten(), 256, [0, 256])

# Plot the histograms
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2,2,2)
plt.plot(hist_orig)
plt.title('Original Image Histogram')

plt.subplot(2,2,3)
plt.imshow(img_eq, cmap='gray')
plt.title('Equalized Image')

plt.subplot(2,2,4)
plt.plot(hist_eq)
plt.title('Equalized Image Histogram')

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img_match, cmap='gray')
plt.title('Matched Image')

plt.subplot(1,2,2)
plt.plot(hist_match)
plt.title('Matched Image Histogram')

plt.show()
```
