# Load libraries and images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
img1 = cv2.imread('/Images/image4.jpg')
img2 = cv2.imread('/Images/image5.jpg')
img3 = cv2.imread('/Images/image6.jpg')
img4 = cv2.imread('/Images/image7.jpg')
```

# A. Average Filter

```python
avg_filter = cv2.blur(img1, (5, 5))

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(avg_filter)
plt.title('Average Filtered Image')
plt.axis('off')
plt.show()
```

# B. Median Filter

```python
median_filter = cv2.medianBlur(img2, 5)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(median_filter)
plt.title('Median Filtered Image')
plt.axis('off')

plt.show()
```

# C. Laplacian Filter

```python
laplacian_filter = cv2.Laplacian(img3, cv2.CV_64F, ksize=5)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(laplacian_filter)
plt.title('Laplacian Filtered Image')
plt.axis('off')

plt.show()
```

# D. Sobel Filter

```python
# Apply the Sobel filter (X direction)
sobel_x_filter = cv2.Sobel(img4, cv2.CV_64F, 1, 0, ksize=7)

# Apply the Sobel filter (Y direction)
sobel_y_filter = cv2.Sobel(img4, cv2.CV_64F, 0, 1, ksize=7)

# Add the filtered images (X and Y direction)
sobel_x_y_filter = cv2.addWeighted(sobel_x_filter, 0.5, sobel_y_filter, 0.5, 0)

plt.figure(figsize=(17,17))
plt.subplot(141)
plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(142)
plt.imshow(sobel_x_filter)
plt.title('Sobel Filtered Image (X direction)')
plt.axis('off')

plt.subplot(143)
plt.imshow(sobel_y_filter)
plt.title('Sobel Filtered Image (Y direction)')
plt.axis('off')

plt.subplot(144)
plt.imshow(sobel_x_y_filter)
plt.title('Sobel Filtered Image (X and Y directions)')
plt.axis('off')

plt.show()
```
